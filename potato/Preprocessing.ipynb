```python
import fitz  # PyMuPDF
import re

def extract_english_text(pdf_path):
    doc = fitz.open(pdf_path)
    full_text = ""

    for page in doc:
        text = page.get_text("text")
        full_text += text + "\n"

    return full_text
```

```python
import re
import nltk
nltk.download("punkt")

from nltk.tokenize import sent_tokenize

import re

def clean_english_text(text):

    # Normalize spaces
    text = re.sub(r"\s+", " ", text)

    # Remove Reprint lines
    text = re.sub(r"Reprint\s*\d{4}-\d{2}", "", text)

    # Remove .indd filenames
    text = re.sub(r"\S+\.indd", "", text)

    # Remove floating similarity scores
    text = re.sub(r"\b\d+\.\d{5,}\b", "", text)

    # Remove standalone numbers
    text = re.sub(r"(?m)^\s*\d+\s*$", "", text)

    # Remove numbering like "7."
    text = re.sub(r"(?m)^\s*\d+\.\s*", "", text)

    # Remove repeated hyphens
    text = re.sub(r"-{2,}", "", text)

    # Remove weird OCR symbols
    text = re.sub(r"[®©•¬_=~]", "", text)

    # Fix space before punctuation
    text = re.sub(r"\s+([.,!?])", r"\1", text)

    # Normalize spaces again
    text = re.sub(r"\s+", " ", text)

    return text.strip()


def split_english_sentences(text):
    return sent_tokenize(text)


def preprocess_english(pdf_path):
    raw = extract_english_text(pdf_path)
    clean = clean_english_text(raw)
    sentences = split_english_sentences(clean)
    
    # Filter bad sentences
    good = []
    for s in sentences:
        if len(s) > 30 and s.endswith(('.', '?', '!')):
            good.append(s.strip())
    
    return good
```

```python
english_sentences = preprocess_english("ncert_6_social.pdf")
with open("english_sentences.txt", "w", encoding="utf-8") as f:
    for s in english_sentences:
        f.write(s + "\n")
```

```python
def extract_hindi_text(pdf_path):
    doc = fitz.open(pdf_path)
    full_text = ""

    for page in doc:
        text = page.get_text("text")
        full_text += text + "\n"

    return full_text
```

```python
import re

import re

def clean_hindi_text(text):

    # Normalize spaces
    text = re.sub(r"\s+", " ", text)

    # Remove Reprint lines
    text = re.sub(r"Reprint\s*\d{4}-\d{2}", "", text)

    # Remove .indd type file names
    text = re.sub(r"\S+\.indd", "", text)

    # Remove floating similarity scores (0.878825...)
    text = re.sub(r"\b\d+\.\d{5,}\b", "", text)

    # Remove standalone numbers (page numbers etc.)
    text = re.sub(r"(?m)^\s*\d+\s*$", "", text)

    # Remove numbering like "1." at start of line
    text = re.sub(r"(?m)^\s*\d+\.\s*", "", text)

    # Remove repeated hyphen separators
    text = re.sub(r"-{2,}", "", text)

    # Remove strange OCR artifacts
    text = re.sub(r"[®©•¬_=~]", "", text)

    # Fix spacing around danda (।)
    text = re.sub(r"\s+।", "।", text)

    # Normalize multiple spaces again
    text = re.sub(r"\s+", " ", text)

    return text.strip()


def split_hindi_sentences(text):
    # Hindi sentences end with ।
    sentences = re.split(r"(?<=[।?!])\s+", text)
    return sentences


def preprocess_hindi(pdf_path):
    raw = extract_hindi_text(pdf_path)
    clean = clean_hindi_text(raw)
    sentences = split_hindi_sentences(clean)

    good = []
    for s in sentences:
        if len(s) > 30 and s.endswith(("।", "?", "!")):
            good.append(s.strip())

    return good
```

```python
hindi_sentences = preprocess_hindi("ncert_6_social_hindi.pdf")
# Save to txt
with open("hindi_sentences.txt", "w", encoding="utf-8") as f:
    for s in hindi_sentences:
        f.write(s + "\n")
```
